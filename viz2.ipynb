```python
import pandas as pd
import altair as alt

# Load the data
df = pd.read_csv('paid leave by country.csv')

# --- Data Cleaning and Preparation (Simplified for this example) ---

# Check for missing values (basic check)
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# --- Create a Combined Top/Bottom DataFrame ---

# Sort by Total Paid Vacation Days
df_sorted = df.sort_values('Total Paid Vacation Days', ascending=False)

# Get the top 20 and bottom 20
top_20 = df_sorted.head(20)
bottom_20 = df_sorted.tail(20)

# Combine them into a single DataFrame for easier charting
combined_df = pd.concat([top_20, bottom_20])

# Add a category to distinguish top and bottom
combined_df['Category'] = ['Top 20'] * 20 + ['Bottom 20'] * 20


# --- Visualization: Sorted Bar Chart of Top/Bottom 20 ---

alt.Chart(combined_df).mark_bar().encode(
    x=alt.X('Total Paid Vacation Days:Q', title='Total Paid Vacation Days'),
    y=alt.Y('Country:N', sort='-x'),  # Sort by Total Paid Vacation Days (descending)
    tooltip=['Country', 'Paid Leave Days', 'Paid Public Holidays', 'Total Paid Vacation Days', 'Category'],
    color=alt.Color('Category:N', scale=alt.Scale(domain=['Top 20', 'Bottom 20'], range=['#1f77b4', '#ff7f0e'])) # Distinct colors
).properties(
    title='Disparity in Total Paid Time Off: Top 20 vs. Bottom 20 Countries',
    width=700,
    height=600
)
```

```python
import pandas as pd
import altair as alt

# Load the data
df = pd.read_csv('paid leave by country.csv')

# --- Data Cleaning and Preparation (Simplified for this example) ---

# Check for missing values (basic check)
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# --- Create a Combined Top/Bottom DataFrame ---

# Sort by Total Paid Vacation Days
df_sorted = df.sort_values('Total Paid Vacation Days', ascending=False)

# Get the top 10 and bottom 10
top_10 = df_sorted.head(10)
bottom_10 = df_sorted.tail(10)

# Combine them into a single DataFrame for easier charting
combined_df = pd.concat([top_10, bottom_10])

# Add a category to distinguish top and bottom
combined_df['Category'] = ['Top 10'] * 10 + ['Bottom 10'] * 10


# --- Visualization: Sorted Bar Chart of Top/Bottom 10 ---

alt.Chart(combined_df).mark_bar().encode(
    x=alt.X('Total Paid Vacation Days:Q', title='Total Paid Vacation Days'),
    y=alt.Y('Country:N', sort='-x'),
    tooltip=['Country', 'Paid Leave Days', 'Paid Public Holidays', 'Total Paid Vacation Days', 'Category'],
    color=alt.Color('Category:N', scale=alt.Scale(scheme='tableau10'))  # Using a named scheme
).properties(
    title='Disparity in Total Paid Time Off: Top 10 vs. Bottom 10 Countries',
    width=700,
    height=500
)
```

```python
import pandas as pd
import altair as alt

# (Rest of the code from the previous examples, up to the chart definition)
df = pd.read_csv('paid leave by country.csv')
df_sorted = df.sort_values('Total Paid Vacation Days', ascending=False)
top_10 = df_sorted.head(10)
bottom_10 = df_sorted.tail(10)
combined_df = pd.concat([top_10, bottom_10])
combined_df['Category'] = ['Top 10'] * 10 + ['Bottom 10'] * 10

alt.Chart(combined_df).mark_bar().encode(
    x=alt.X('Total Paid Vacation Days:Q', title='Total Paid Vacation Days', axis=alt.Axis(grid=False)),
    y=alt.Y('Country:N', sort='-x', axis=alt.Axis(
        gridColor='lightgray',
        gridWidth=0.5,
        gridDash=[4, 2],
        gridOpacity=0.5
    )),
    tooltip=['Country', 'Paid Leave Days', 'Paid Public Holidays', 'Total Paid Vacation Days', 'Category'],
    color=alt.Color('Category:N', scale=alt.Scale(scheme='tableau10'), legend=None)  # Remove the legend
).properties(
    title='Disparity in Total Paid Time Off: Top 10 vs. Bottom 10 Countries',
    width=700,
    height=500
)

```
