```python
import pandas as pd
```

```python
data_folder = 'data/'
dish_df = pd.read_csv(data_folder + 'Dish.csv')
menu_df = pd.read_csv(data_folder + 'Menu.csv')
menu_item_df = pd.read_csv(data_folder + 'MenuItem.csv')
menu_page_df = pd.read_csv(data_folder + 'MenuPage.csv')
```

```python
# check missing values
def missing_values(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    return pd.concat([total, percent], axis=1, keys=['Total Missing Rows', 'Percent'])
```

```python
missing_values(dish_df)
```

```python
missing_values(menu_df)
```

```python
missing_values(menu_item_df)
```

```python
missing_values(menu_page_df)
```

```python
# delete column with > 60% missing values
def del_missing(df, length, threshold=0.4):
    return df.dropna(thresh=length * threshold, axis=1)
```

```python
dish_df_clean = del_missing(dish_df, len(dish_df))
```

```python
menu_df_clean = del_missing(menu_df, len(menu_df))
```

```python
menu_item_df_clean = del_missing(menu_item_df, len(menu_item_df))
```

```python
menu_page_df_clean = del_missing(menu_page_df, len(menu_page_df))
```

```python
print(list(dish_df_clean.columns))
print(list(menu_df_clean.columns))
print(list(menu_item_df_clean.columns))
print(list(menu_page_df_clean.columns))
```

Unnecessary columns
Dish: None
Menu: sponsor, notes, call_number, status, place
MenuItem: updated_at
MenuPage: image_id, uuid

```python
# drop unnecessary columns
menu_df_clean = menu_df_clean.drop(['sponsor', 'notes', 'call_number', 'status', 'place'], axis=1)
menu_item_df_clean = menu_item_df_clean.drop(['updated_at'], axis=1)
menu_page_df_clean = menu_page_df_clean.drop(['image_id', 'uuid'], axis=1)
print("Columns from Dish: ", list(dish_df_clean.columns))
print("Columns from Menu: ", list(menu_df_clean.columns))
print("Columns from MenuItem: ", list(menu_item_df_clean.columns))
print("Columns from MenuPage: ", list(menu_page_df_clean.columns))
```

Must have value columns:
Dish: name
Menu: event, venue
MenuItem: price, dish_id, created_at
MenuPage: None

```python
# delete rows with missing values in must have columns
dish_df_clean = dish_df_clean.dropna(subset=['name'])
menu_df_clean = menu_df_clean.dropna(subset=['event', 'venue'])
menu_item_df_clean = menu_item_df_clean.dropna(subset=['price', 'dish_id', 'created_at'])
```
