```python
import pandas as pd
```

# Data Cleaning

```python
data_folder = 'data/'

df_raw = pd.read_csv(data_folder + 'ScreenTime_Data.CSV', header=0)
pd.DataFrame(df_raw.columns)
```

```python
# Drop rows with Included != 1, ASQ_Included != 1, Educational_Content_Included != 1
df = df_raw[(df_raw['Included'] == 1) & (df_raw['ASQ_Included'] == 1) & (df_raw['Educational_Content_Included'] == 1)]
df
```

select columns:
- EDUCATIONAL_Interactive in name
- ASQ in name
- AVG_ALLWEEK_Continuous_EDUCATIONAL in name
- MOSTOFTEN not in name
- CBCL in name
- exactly: Child_Age
- exactly: Gender
- exactly: Years_Education
- exactly: Caregiver_Relationship_to_child

```python
df = df[[
    col for col in df.columns
    if ('EDUCATIONAL_Interactive' in col and 'MOSTOFTEN' not in col) or
       ('ASQ' in col) or
       ('CBCL' in col) or
       ('AVG_ALLWEEK_Continuous_EDUCATIONAL' in col) or
       (col in ['Child_Age', 'Gender', 'Years_Education', 'Caregiver_Relationship_to_child', 'PARTICIPANT - NOTE DUPLICATES (1ST ROW RAW DATA, 2ND ROW CODED)'])
]]
df.columns
```

```python
# Rename columns
# all lowercase, symbols(space, /, ') to underscores
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('/', '_').str.replace("'", '')
# avg_allweek_continuous_educational_ -> week_
# educational_interactive_solo_ -> is_solo_
df.columns = df.columns.str.replace('avg_allweek_continuous_educational_', 'week_')
df.columns = df.columns.str.replace('educational_interactive_solo_', 'is_solo_')
# PARTICIPANT - NOTE DUPLICATES (1ST ROW RAW DATA, 2ND ROW CODED) -> id
df = df.rename(columns={'participant_-_note_duplicates_(1st_row_raw_data,_2nd_row_coded)': 'id'})
df.columns
```

```python
week_hours_pattern = "week_"
interactive_pattern = "is_solo_"
# Interactive columns: 1 = with someone; 2 = alone; 3 = does not engage with the content
interactive_columns = [col for col in df.columns if interactive_pattern in col]

df['interactive'] = 0.0
df['solo'] = 0.0

for solo_col in interactive_columns:
    category = solo_col.replace(interactive_pattern, '')
    hours_col = week_hours_pattern + category
    for index, row in df.iterrows():
        interaction_type = row[solo_col]
        if interaction_type == 1:  # Interactive
          df.loc[index, 'interactive'] += row[hours_col]
        elif interaction_type == 2:  # Solo
          df.loc[index, 'solo'] += row[hours_col]
df[['solo', 'interactive']]
```

```python
# Scale scores: min-max to 1-100, int

# drop rows with missing values in 'asq_communication_all_children', 'asq_prob_sol_avg_score', 'cbcl_attention_problems_t_score'

df = df.dropna(subset=['asq_communication_all_children', 'asq_prob_sol_total_score', 'cbcl_attention_problems_t_score'])

min_comm = df['asq_communication_all_children'].min()
max_comm = df['asq_communication_all_children'].max()

min_prob = df['asq_prob_sol_total_score'].min()
max_prob = df['asq_prob_sol_total_score'].max()

min_att = df['cbcl_attention_problems_t_score'].min()
max_att = df['cbcl_attention_problems_t_score'].max()

# Apply min-max scaling and convert to integer
df['communication_score'] = ((df['asq_communication_all_children'] - min_comm) / (max_comm - min_comm) * 99 + 1).astype(int)
df['problem_solving_score'] = ((df['asq_prob_sol_total_score'] - min_prob) / (max_prob - min_prob) * 99 + 1).astype(int)
df['attention_score'] = ((df['cbcl_attention_problems_t_score'] - min_att) / (max_att - min_att) * 99 + 1).astype(int)

# --- 2. Calculate avg_score ---

df['avg_score'] = ((df['communication_score'] + df['problem_solving_score'] + df['attention_score']) / 3).astype(int)

df[['communication_score', 'problem_solving_score', 'attention_score', 'avg_score']]
```

```python
# turn is_solo_? to int
interactive_columns = [col for col in df.columns if interactive_pattern in col]
for col in interactive_columns:
    df[col] = df[col].astype(int)

df.dtypes
```

```python
def pivot_activity_data(df: pd.DataFrame) -> pd.DataFrame:
    activity_types = [
        "documentaries_how_tos",
        "creative_programs",
        "learning_apps",
        "information_searching",
        "writing_drawing_painting",
        "taking_photos_filming",
    ]
    
    # Melt to get 'spent'
    df_spent = df.melt(
        id_vars=[
            "id",
            "child_age",
            "gender",
            "caregiver_relationship_to_child",
            "years_education",
            "communication_score",
            "problem_solving_score",
            "attention_score",
            "avg_score",
        ],
        value_vars=[f"week_{x}" for x in activity_types],
        var_name="activity_type",
        value_name="spent",
    )
    df_spent["activity_type"] = df_spent["activity_type"].str.replace("week_", "", regex=False)

    # Melt to get values for 'solo_interactive_type'
    df_solo = df.melt(
        id_vars=[
            "id",
            "child_age",
            "gender",
            "caregiver_relationship_to_child",
            "years_education",
            "communication_score",
            "problem_solving_score",
            "attention_score",
            "avg_score",
        ],
        value_vars=[f"is_solo_{x}" for x in activity_types],
        var_name="temp_activity_type",
        value_name="solo_value",
    )
    df_solo["temp_activity_type"] = df_solo["temp_activity_type"].str.replace(
        "is_solo_", "", regex=False
    )

    # Merge the two melted DataFrames
    df_merged = pd.merge(
        df_spent,
        df_solo,
        left_on=[
            "id",
            "child_age",
            "gender",
            "caregiver_relationship_to_child",
            "years_education",
            "communication_score",
            "problem_solving_score",
            "attention_score",
            "avg_score",
            "activity_type",
        ],
        right_on=[
            "id",
            "child_age",
            "gender",
            "caregiver_relationship_to_child",
            "years_education",
            "communication_score",
            "problem_solving_score",
            "attention_score",
            "avg_score",
            "temp_activity_type",
        ],
        how="inner",
    )

    # Map solo/interactive integer values to labels
    df_merged["solo_interactive_type"] = df_merged["solo_value"].map({1: "Interactive", 2: "Solo", 3: None})

    # Select and order final columns
    df_final = df_merged[
        [
            "id",
            "child_age",
            "gender",
            "caregiver_relationship_to_child",
            "years_education",
            "activity_type",
            "spent",
            "solo_interactive_type",
            "communication_score",
            "problem_solving_score",
            "attention_score",
            "avg_score",
        ]
    ]

    return df_final

new_df = pivot_activity_data(df)
new_df.columns
```
